# Predicting whether a credit card transaction is fraudulent or not

```python
# 1. Import Data
import pandas as pd

# Load your dataset
df = pd.read_csv("creditcard.csv")   
df.head()
```

# 3. Data Cleaning / Data Preprocessing

```python
df.drop_duplicates(inplace=True)
```

```python
print("Shape:", df.shape)
```

# 4. EDA / Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(df)
plt.show()
```

```python
plt.boxplot(df['Amount'])
plt.show()
```

```python
class_counts = df['Class'].value_counts()
class_counts
```

```python
plt.bar(class_counts.index, class_counts.values, color=['green', 'red'])
plt.xticks([0, 1], ['Not Fraud', 'Fraud'])
plt.title("Class Distribution")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
# Note: Dataset is highly imbalanced
# 0 â†’ Not Fraud (~2 lakh), 1 â†’ Fraud (~400)
# If we directly train our model on this dataset, 
# it will always predict 'Not Fraud' because the data is highly imbalanced. 
# Even if the model shows 99% accuracy, it will still be a useless model â€” that's why we balanced the dataset.
```

```python
fraud = df[df['Class'] == 1]
not_fraud = df[df['Class'] == 0]
```

```python
not_fraud_sampled = not_fraud.sample(n=len(fraud)*4, random_state=42)
```

```python
counts = [len(not_fraud_sampled), len(fraud)]
labels = ['Not Fraud', 'Fraud']

plt.bar(labels, counts, color=['green', 'red'])
plt.title("Class Distribution")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
# adding the dataset
df_balanced = pd.concat([fraud, not_fraud_sampled])
```

```python
df_balanced = df_balanced.sample(frac=1, random_state=42)
```

```python
# Correlation matrix on balanced dataset
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = df_balanced.corr()
```

```python
# Plot the heatmap
plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix (Balanced Dataset)", fontsize=16)
plt.show()
```

# Splitting the dataset

```python
X = df_balanced.drop('Class', axis=1)
y = df_balanced['Class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42)
```

# Using svm model

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

svm = SVC(kernel='rbf', class_weight='balanced')  # 'balanced' handles class imbalance
svm.fit(X_train, y_train)
```

```python

# Predict
y_pred = svm.predict(X_test)
# Use the original test features (X_test), and test labels (y_test)
print("Accuracy Score:", svm.score(X_test, y_test))

# Evaluate
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# ROC AUC score
roc = roc_auc_score(y_test, y_pred)
print("ROC AUC Score:", roc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Fraud', 'Fraud'],
            yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred_proba = rf.predict_proba(X_test)[:, 1]

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# ROC AUC Score
roc = roc_auc_score(y_test, y_pred_proba)
print("ROC AUC Score:", roc)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Fraud', 'Fraud'],
            yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest Confusion Matrix")
plt.show()
```

# XGBoost


```python
import xgboost as xgb
```

```python
xgboostModel = xgb.XGBClassifier(
    objective='binary:logistic',
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=len(not_fraud)/len(fraud),  # balance fraud class
    random_state=42
)

xgboostModel.fit(X_train, y_train)
```

```python
y_pred = xgboostModel.predict(X_test)
y_pred_proba = xgboostModel.predict_proba(X_test)[:, 1]

# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# ROC AUC Score
roc = roc_auc_score(y_test, y_pred_proba)
print("ROC AUC Score:", roc)
```

```python
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Not Fraud', 'Fraud'],
            yticklabels=['Not Fraud', 'Fraud'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("XGBoost Confusion Matrix")
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```
